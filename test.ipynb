```python
from data import load_data
import matplotlib.pyplot as plt

import random
```

```python
nb_x, nb_y = 3, 3

figs, axes = plt.subplots(nb_x, nb_y)

chosen_idxs = random.sample(range(len(ds)), nb_x * nb_y)

for i in range(nb_x):
    x = i * nb_x
    for j in range(nb_y):
        idx = x + j
        ds_idx = chosen_idxs[idx]

        axes[i, j].imshow(ds[ds_idx][0])
        axes[i, j].set_title(idx2label[ds[ds_idx][1]])
        axes[i, j].axis('off')
plt.show()
```
